# Cardinalidad

**Por ejemplo:** El conjunto  **A={a, b, c}** tiene tres elementos, por lo cual su cardinal o tama√±o es ese n√∫mero:

El conjunto **B={1, 2, 1}** tiene dos elementos diferentes, pues el **"1"** se escribi√≥ dos veces, por lo tanto **|B|=2.**
