- CÃ¡lculo de la imagen de la funciÃ³n

Para calcular la imagen de la funciÃ³n sobre un dominio especÃ­fico, podemos aplicar la funciÃ³n a cada elemento del dominio:

Para ilustrar el concepto de imagen y preimagen en Python, asÃ­ como algunos principios bÃ¡sicos de demostraciÃ³n, podemos escribir un pequeÃ±o script que implemente una funciÃ³n y luego calcule su imagen y preimagen para ciertos conjuntos. Vamos a crear una funciÃ³n cuadrÃ¡tica y demostrar algunas de sus propiedades.



### DefiniciÃ³n de la funciÃ³n cuadrÃ¡tica

Primero, definamos una funciÃ³n cuadrÃ¡tica simple \( f(x) = x^2 \):

```python
def f(x):
    return x**2
```

#### CÃ¡lculo de la imagen de la funciÃ³n

Para calcular la imagen de la funciÃ³n sobre un dominio especÃ­fico, podemos aplicar la funciÃ³n a cada elemento del dominio:

```python
def calcular_imagen(func, dominio):
    return [func(x) for x in dominio]

# Ejemplo de uso
dominio = list(range(-10, 11))  # Dominio de -10 a 10
imagen = calcular_imagen(f, dominio)
print("Imagen de f(x) sobre el dominio:", imagen)
```

#### CÃ¡lculo de la preimagen de un conjunto

Para encontrar la preimagen de un conjunto \( C \), buscamos todos los valores en el dominio que, al ser aplicados a la funciÃ³n, resulten en elementos de \( C \):

```python
def calcular_preimagen(func, codominio, dominio):
    preimagen = []
    for y in codominio:
        preimagen.extend([x for x in dominio if func(x) == y])
    return preimagen

# Ejemplo de uso
codominio = [1, 4, 9, 16, 25]  # Valores para los que queremos calcular la preimagen
preimagen = calcular_preimagen(f, codominio, dominio)
print("Preimagen de C bajo f(x):", preimagen)
```

### Principios de DemostraciÃ³n en Python

#### DemostraciÃ³n de inyectividad

Para verificar si una funciÃ³n es inyectiva en un dominio dado, podemos comprobar si cada valor de la imagen es Ãºnico:

```python
def es_inyectiva(func, dominio):
    imagen = calcular_imagen(func, dominio)
    return len(imagen) == len(set(imagen))

# Ejemplo de uso
es_inyectiva_f = es_inyectiva(f, dominio)
print("Â¿Es f(x) inyectiva sobre el dominio?", es_inyectiva_f)
```

#### DemostraciÃ³n por inducciÃ³n (nÃºmeros naturales)

Para demostrar que la suma de los primeros \( n \) nÃºmeros naturales es \( \frac{n(n+1)}{2} \), podemos escribir una funciÃ³n que calcule esta suma y compararla con la fÃ³rmula:

```python
def suma_numeros_naturales(n):
    return n * (n + 1) // 2

def suma_directa(n):
    return sum(range(1, n + 1))

# Ejemplo de uso
n = 10
print("Suma usando la fÃ³rmula:", suma_numeros_naturales(n))
print("Suma directa:", suma_directa(n))
print("Â¿Son iguales?", suma_numeros_naturales(n) == suma_directa(n))
```

### Script completo

AquÃ­ tienes el script completo que implementa todos los conceptos mencionados:

```python
def f(x):
    return x**2

def calcular_imagen(func, dominio):
    return [func(x) for x in dominio]

def calcular_preimagen(func, codominio, dominio):
    preimagen = []
    for y in codominio:
        preimagen.extend([x for x in dominio if func(x) == y])
    return preimagen

def es_inyectiva(func, dominio):
    imagen = calcular_imagen(func, dominio)
    return len(imagen) == len(set(imagen))

def suma_numeros_naturales(n):
    return n * (n + 1) // 2

def suma_directa(n):
    return sum(range(1, n + 1))

# Ejemplo de uso
dominio = list(range(-10, 11))  # Dominio de -10 a 10
imagen = calcular_imagen(f, dominio)
print("Imagen de f(x) sobre el dominio:", imagen)

codominio = [1, 4, 9, 16, 25]  # Valores para los que queremos calcular la preimagen
preimagen = calcular_preimagen(f, codominio, dominio)
print("Preimagen de C bajo f(x):", preimagen)

es_inyectiva_f = es_inyectiva(f, dominio)
print("Â¿Es f(x) inyectiva sobre el dominio?", es_inyectiva_f)

n = 10
print("Suma usando la fÃ³rmula:", suma_numeros_naturales(n))
print("Suma directa:", suma_directa(n))
print("Â¿Son iguales?", suma_numeros_naturales(n) == suma_directa(n))
```
