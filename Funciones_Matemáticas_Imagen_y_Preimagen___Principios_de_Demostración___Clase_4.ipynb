- Cálculo de la imagen de la función

Para calcular la imagen de la función sobre un dominio específico, podemos aplicar la función a cada elemento del dominio:

Para ilustrar el concepto de imagen y preimagen en Python, así como algunos principios básicos de demostración, podemos escribir un pequeño script que implemente una función y luego calcule su imagen y preimagen para ciertos conjuntos. Vamos a crear una función cuadrática y demostrar algunas de sus propiedades.



### Definición de la función cuadrática

Primero, definamos una función cuadrática simple \( f(x) = x^2 \):

```python
def f(x):
    return x**2
```

#### Cálculo de la imagen de la función

Para calcular la imagen de la función sobre un dominio específico, podemos aplicar la función a cada elemento del dominio:

```python
def calcular_imagen(func, dominio):
    return [func(x) for x in dominio]

# Ejemplo de uso
dominio = list(range(-10, 11))  # Dominio de -10 a 10
imagen = calcular_imagen(f, dominio)
print("Imagen de f(x) sobre el dominio:", imagen)
```

#### Cálculo de la preimagen de un conjunto

Para encontrar la preimagen de un conjunto \( C \), buscamos todos los valores en el dominio que, al ser aplicados a la función, resulten en elementos de \( C \):

```python
def calcular_preimagen(func, codominio, dominio):
    preimagen = []
    for y in codominio:
        preimagen.extend([x for x in dominio if func(x) == y])
    return preimagen

# Ejemplo de uso
codominio = [1, 4, 9, 16, 25]  # Valores para los que queremos calcular la preimagen
preimagen = calcular_preimagen(f, codominio, dominio)
print("Preimagen de C bajo f(x):", preimagen)
```

### Principios de Demostración en Python

#### Demostración de inyectividad

Para verificar si una función es inyectiva en un dominio dado, podemos comprobar si cada valor de la imagen es único:

```python
def es_inyectiva(func, dominio):
    imagen = calcular_imagen(func, dominio)
    return len(imagen) == len(set(imagen))

# Ejemplo de uso
es_inyectiva_f = es_inyectiva(f, dominio)
print("¿Es f(x) inyectiva sobre el dominio?", es_inyectiva_f)
```

#### Demostración por inducción (números naturales)

Para demostrar que la suma de los primeros \( n \) números naturales es \( \frac{n(n+1)}{2} \), podemos escribir una función que calcule esta suma y compararla con la fórmula:

```python
def suma_numeros_naturales(n):
    return n * (n + 1) // 2

def suma_directa(n):
    return sum(range(1, n + 1))

# Ejemplo de uso
n = 10
print("Suma usando la fórmula:", suma_numeros_naturales(n))
print("Suma directa:", suma_directa(n))
print("¿Son iguales?", suma_numeros_naturales(n) == suma_directa(n))
```

### Script completo

Aquí tienes el script completo que implementa todos los conceptos mencionados:

```python
def f(x):
    return x**2

def calcular_imagen(func, dominio):
    return [func(x) for x in dominio]

def calcular_preimagen(func, codominio, dominio):
    preimagen = []
    for y in codominio:
        preimagen.extend([x for x in dominio if func(x) == y])
    return preimagen

def es_inyectiva(func, dominio):
    imagen = calcular_imagen(func, dominio)
    return len(imagen) == len(set(imagen))

def suma_numeros_naturales(n):
    return n * (n + 1) // 2

def suma_directa(n):
    return sum(range(1, n + 1))

# Ejemplo de uso
dominio = list(range(-10, 11))  # Dominio de -10 a 10
imagen = calcular_imagen(f, dominio)
print("Imagen de f(x) sobre el dominio:", imagen)

codominio = [1, 4, 9, 16, 25]  # Valores para los que queremos calcular la preimagen
preimagen = calcular_preimagen(f, codominio, dominio)
print("Preimagen de C bajo f(x):", preimagen)

es_inyectiva_f = es_inyectiva(f, dominio)
print("¿Es f(x) inyectiva sobre el dominio?", es_inyectiva_f)

n = 10
print("Suma usando la fórmula:", suma_numeros_naturales(n))
print("Suma directa:", suma_directa(n))
print("¿Son iguales?", suma_numeros_naturales(n) == suma_directa(n))
```

Este script demuestra cómo calcular la imagen y la preimagen de una función, verificar la inyectividad de una función, y demostrar la fórmula de la suma de los primeros \( n \) números naturales mediante programación en Python.
