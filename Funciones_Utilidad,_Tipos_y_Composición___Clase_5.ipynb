### Definici√≥n de Funci√≥n en Matem√°ticas:
En matem√°ticas, una funci√≥n es una relaci√≥n entre un conjunto de entradas (llamado dominio) y un conjunto de salidas (llamado codominio), donde cada elemento del dominio se asigna exactamente a un elemento del codominio. Formalmente, si tienes una funci√≥n \( f \), se escribe como \( f: A \rightarrow B \), donde \( A \) es el dominio y \( B \) es el codominio.

### Utilidad de las Funciones en Matem√°ticas:
1. **Modelado de Relaciones:** Permiten modelar relaciones entre variables y fen√≥menos matem√°ticos.
2. **Resoluci√≥n de Problemas:** Ayudan a resolver ecuaciones y problemas matem√°ticos de manera estructurada.
3. **Generalizaci√≥n y Abstracci√≥n:** Permite generalizar patrones matem√°ticos y abstraer conceptos clave.

### Tipos de Funciones en Matem√°ticas:
1. **Funciones Lineales:**
   - Tienen la forma \( f(x) = ax + b \), donde \( a \) y \( b \) son constantes.
   - Ejemplo: \( f(x) = 2x + 1 \).

2. **Funciones Cuadr√°ticas:**
   - Tienen la forma \( f(x) = ax^2 + bx + c \), donde \( a \), \( b \) y \( c \) son constantes y \( a \neq 0 \).
   - Ejemplo: \( f(x) = x^2 - 3x + 2 \).

3. **Funciones Trigonom√©tricas:**
   - Incluyen funciones como seno, coseno y tangente, definidas en t√©rminos de √°ngulos.
   - Ejemplo: \( f(x) = \sin(x) \).

### Composici√≥n de Funciones en Matem√°ticas:
La composici√≥n de funciones en matem√°ticas implica aplicar una funci√≥n sobre los resultados de otra funci√≥n. Si tienes dos funciones \( f \) y \( g \):
- La composici√≥n se denota como \( (f \circ g)(x) = f(g(x)) \).
- Ejemplo: Si \( f(x) = 2x \) y \( g(x) = x + 1 \), entonces \( (f \circ g)(x) = f(g(x)) = 2(x + 1) = 2x + 2 \).

### Ejemplos de Funciones en Python:

```python
import math

# Funci√≥n lineal
def funcion_lineal(x):
    return 2 * x + 3

# Funci√≥n cuadr√°tica
def funcion_cuadratica(x):
    return x**2 - 4 * x + 4

# Funci√≥n trigonom√©trica (seno)
def funcion_trigonometrica(x):
    return math.sin(x)

# Funci√≥n exponencial
def funcion_exponencial(x):
    return 2 ** x

# Ejemplos de evaluaci√≥n
resultado1 = funcion_lineal(1)
resultado2 = funcion_cuadratica(2)
resultado3 = funcion_trigonometrica(math.pi / 2)
resultado4 = funcion_exponencial(3)

# Imprimir resultados
print(f"Funci√≥n lineal: f(1) = {resultado1}")
print(f"Funci√≥n cuadr√°tica: f(2) = {resultado2}")
print(f"Funci√≥n trigonom√©trica (seno): f(pi/2) = {resultado3}")
print(f"Funci√≥n exponencial: f(3) = {resultado4}")
```

Este c√≥digo define todas las funciones (lineal, cuadr√°tica, trigonom√©trica y exponencial) y luego las eval√∫a con ejemplos espec√≠ficos.
